## Hypotheses Test
### Objective
- How Product category impact sales
### Hypotheses
- Null Hypothesis(Ho)- Product category does not affect sales
- Alternate Hypothesis(H1)- Product category  affects sales

```python
categories = df['Product Category'].unique()
nets = [df[df['Product Category'] == i]['Net Amount'] for i in categories]
```

```python
from scipy.stats import f_oneway
f_stat, p_val = f_oneway(*nets)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")
```

```python
alpha = 0.05
if p_val < alpha:
    print("Reject Null Hypothesis  -> Product category significantly affects sales")
else:
    print("Fail to Reject Null Hypothesis  -> Product category does not significantly affect sales")
```

### Insights 
- Statistical testing shows no significant difference in average Net Amount across categories (p > 0.05).
- However, total revenue varies significantly due to different transaction volumes, with Electronics and Clothing contributing the most.

## Hypotheses Test
### Objective
- How does discount impact sales
### Hypotheses
- Null Hypothesis(Ho)- Discount Does not Impact sales
- Alternate Hypothesis(H1)- Discount  Impacts sales

```python
Discounted=df[df["Discount Availed"]=="Yes"]["Net Amount"]
Not_Discounted=df[df["Discount Availed"]=="No"]["Net Amount"]
```

```python
from scipy.stats import ttest_ind
t_stat,p_value=ttest_ind(Discounted,Not_Discounted)
print("p_value:",p_value)
```

```python
if p_value< 0.05:
    print(" reject Null Hypothesis(Ho): Discount Impacts sales ")
else:
     print(" fail to reject Null Hypothesis(Ho): Discount Does not Impact sales")
```

```python
d = {"Discount Amount (INR)": "mean",  "Gross Amount": "sum","CID": "count"}
result =df.groupby(["Product Category","Discount Availed"]).agg(d)
result = result.reset_index()
result['Gross Amount'] = result['Gross Amount'].astype(int)
print(result)
```

### Insights 
- High-revenue categories are Electronics, Clothing, and Beauty & Health, contributing most to total sales.
- Discounts slightly increase revenue for mid-range categories like Clothing and Home & Kitchen.
- High-ticket categories like Electronics and Pet Care generate slightly more revenue without discounts.
- Average discount is around 270–280 INR and consistent across all categories.
### Recommendations 
- Focus marketing and inventory on high-revenue categories to maximize overall sales.
- Offer discounts selectively for mid-range or impulse-buy categories to boost transactions.
- Reduce or optimize discounts for high-value categories to protect margins.
- Monitor revenue vs discount effectiveness regularly to refine pricing and promotion strategies.
